```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\myaha\OneDrive\Desktop\Jupyter Notebook files\CN Project\task_1.csv')
```

```python
df.head()
```

```python
indata = df["Input Data"]
```

```python
outdata = df["Output Data"]
```

```python
process_delay = (indata*10**6)/(32*10**6)
process_delay.head()
```

```python
input_delay = (indata*10**3)/(75*10**6)
output_delay = (outdata*10**3)/(75*10**6)
```

```python
input_delay.head()
```

```python
output_delay.head()
```

```python
final_delay = input_delay + output_delay + process_delay
final_delay
```

```python
def FCFS(x):
    sum=0
    fcfs = np.empty((0))
    for i in range(len(x)):
        sum = sum + final_delay[i]
        fcfs = np.append(fcfs,sum)
    return fcfs
```

```python
fcfs = FCFS(final_delay)
fcfs
```

```python
task_number = np.arange(0, 700, 1)
#task_number
```

```python
# FCFS Display
plt.plot(task_number,fcfs)
plt.xlabel("Tasks")
plt.ylabel("Delay Time")
plt.title("FCFS")
plt.show()
```

```python
# SJF Start
temp = final_delay.to_numpy()
temp = np.sort(temp)
temp
```

```python
sum=0
sjf = np.empty((0))
sjf_t = np.empty((0))
for i in range(len(temp)):
    sum=sum + temp[i]
    sjf_t = np.append(sjf_t,temp[i])
    sjf = np.append(sjf,sum)
sjf
```

```python
plt.plot(task_number,sjf,label = "SJF")
plt.xlabel("Tasks")
plt.ylabel("Delay Time")
plt.title("SJF")
```

```python
# Priority Scheduling
p = df
p = p.sort_values("Priority Number")
p.head()
```

```python
# input output and proccess delay calculation
p_indata = p["Input Data"]
p_outdata = p["Output Data"]
p_process_delay = (p_indata*10**6)/(32*10**6)
p_input_delay = (p_indata*10**3)/(75*10**6)
p_output_delay = (p_outdata*10**3)/(75*10**6)
p_final_delay = p_input_delay + p_output_delay + p_process_delay
p_final_delay
```

```python
t_pr = p_final_delay.to_numpy()
t_pr
```

```python
# Apply Priority Sceduling
sum=0
pr = np.empty((0))
for i in range(len(t_pr)):
    sum=sum + t_pr[i]
    pr = np.append(pr,sum)
pr
```

```python
plt.plot(task_number,pr,label = "Priority Sceduling")
plt.xlabel("Tasks")
plt.ylabel("Delay Time")
plt.title("Priority Scheduding")
```

```python
plt.plot(task_number,sjf,label = "SJF")
plt.plot(task_number,fcfs,label = "FCFS")
plt.plot(task_number,pr,label = "Priority Sceduling")
plt.xlabel("Task")
plt.ylabel("Delay Time")
plt.title("Comparison")
plt.legend()
```
